```python
import pandas as pd
import numpy as np
from pathlib import Path
root = Path('..')
out = root/'output'
```

```python
a = pd.read_csv(out/'push_triple_w145.csv').sort_values('Index')
b = pd.read_csv(out/'v3adv_lgbm.csv').sort_values('Index')
m = a.merge(b, on='Index', suffixes=('_stack','_lg'))
```

```python
base = pd.read_csv(out/'v3adv_w145lgb_0983.csv').sort_values('Index').demand.values
lg = m.demand_lg.values
# tail fix
m2 = lg >= np.quantile(lg, 0.996)
final = base.copy()
final[m2] = np.maximum(final[m2], 0.97*lg[m2])
final = np.clip(final,0,1)
```

```python
pd.DataFrame({'Index':m.Index.astype(int),'demand':final}).to_csv(out/'submission.csv', index=False)
print('mean', final.mean())
```
